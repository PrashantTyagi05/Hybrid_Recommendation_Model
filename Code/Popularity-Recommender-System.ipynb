```python
import numpy as np
import pandas as pd
```

```python
m = df['vote_count'].quantile(0.9) # min number of votes to qualify as recommended movie.
C = df['vote_average'].mean()      # average voting in iMdb
```

```python
print(m) , print(C)
```

*Applying the iMdb formula to rate the movies or the recommendation value based on popular votes.*

```python
def weighted(x , m=m , C=C):
    v = x['vote_count']
    R = x['vote_average']
    
    return (v/(v+m))*R + (m/(v+m))*C
```

```python
apply_filter = df.loc[df['vote_count'] >=m]
```

```python
df.shape , apply_filter.shape
```

```python
apply_filter['score'] = apply_filter.apply(weighted , axis=1)
```

```python
apply_filter = apply_filter.sort_values('score' , ascending=False)
```

```python
def genre_based_popularity(genre):
    mask = apply_filter.genres.apply(lambda x: genre in x)
    filtered_movie = apply_filter[mask]
    filtered_movie = filtered_movie.sort_values(by='score', ascending=False)
    return filtered_movie
```

```python
def director_based_popularity(director):
    df = popularity_data[popularity_data['director']==director]
    filtered_movie = df.sort_values(by='score', ascending=False)
    return filtered_movie

director_based_popularity('christophernolan')[['title', 'score']].head(25)
```
